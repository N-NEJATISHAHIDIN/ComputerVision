```python
# Solution: Compute oriented, scaled image patch

## Solution
def get_image_patch(image, patch_center_coord, patch_half_width=16, orientation=0.0, scale=1.0):
    """Gets a scaled, rotated, patch from a reference image.
    Arguments:
    - image: a 2D image
    - patch_center_coord: a 2-element tuple corresponding to the patch center in the image.
    - patch_width: the width of the output patch.
    - orientation: the rotation of the patch with respect to the original image.
    - scale: the spacing between pixels in the output relative to the input.
    
    If the orientation=0 and scale=1, this function should be equivalent to simply
    computing the patch via slicing: patch = image[ix-w:ix+w+1, iy-w:iy+w+1]
    """
    x = np.arange(image.shape[1]).astype(np.float)
    y = np.arange(image.shape[0]).astype(np.float)
    
    # The patch has a different range from the original image.
    patch_width = 2*patch_half_width+1
    xi = np.linspace(-2, 2, patch_width)
    yi = np.linspace(-2, 2, patch_width)

    # Apply the transformations to get the new coordinates
    # (Again, we flip the coordinates due to the convention difference
    # between [rows, columns] and [x, y].)
    cth = np.cos(-orientation)
    sth = np.sin(-orientation)
    transformation_matrix = np.array([
        [cth/scale, sth/scale, patch_center_coord[1]],
        [-sth/scale, cth/scale, patch_center_coord[0]],
        [0, 0, 1]
    ])

    # Perform the interpolation
    patch = np.zeros((patch_width, patch_width))
    image_fn = scipy.interpolate.interp2d(x, y, image, kind='linear', fill_value=0)
    for ii in range(len(xi)):
        for jj in range(len(yi)):
            ct = np.matmul(transformation_matrix, 
                           [[xi[ii]], [yi[jj]], [1]])
            new_x = ct[0]/ct[2]
            new_y = ct[1]/ct[2]
            patch[jj, ii] = image_fn(new_x, new_y)
    
    return patch

def get_descriptor_patch(image, floc, fscale):
    raise NotImplementedError()
```

```python
# Solution: Multi-scale blob detection
import matplotlib.pyplot as plt
import scipy.ndimage
import scipy.ndimage.filters as filters

def get_circ_image(image_size, radius):
    """Create an image of width `image_size` with a circle 
    of radius `radius` in its center."""
    assert(image_size % 2 == 1)
    h = (image_size + 1)//2
    d = np.arange(h)
    d = np.concatenate((d[::-1], d[1:]))
    d = d[:, np.newaxis]
    d_sq = d**2 + d.T ** 2
    # Threshold by squared radius
    d_sq = (d_sq <= radius**2).astype(np.float)
    return d_sq

def get_square_image(image_size, width):
    """Create an image of width `image_size` with a circle 
    of radius `radius` in its center."""
    assert(image_size % 2 == 1)
    h = (image_size + 1)//2
    d = np.arange(h)
    d = np.concatenate((d[::-1], d[1:]))
    d = d[:, np.newaxis]
    d = (d <= width/2).astype(float)
    d_sq = d * d.T
    return d_sq

def get_gaussian_filter(n, sigma=1):
    n = 2 * (n//2) + 1
    h = (n + 1)//2
    d = np.arange(h)
    d = np.concatenate((d[::-1], d[1:]))
    d = d[:, np.newaxis]
    d_sq = d**2 + d.T ** 2
    # Take the gaussian
    g = np.exp(-d_sq/2/(sigma**2))
    # Normalize
    g = g/g.sum().sum()
    return g

def get_LoG_filter(kernel_size, sigma):
    kernel_size = 2 * (kernel_size//2) + 1
    h = (kernel_size + 1)//2
    d = np.arange(h)
    d = np.concatenate((d[::-1], d[1:]))
    d = d[:, np.newaxis]
    d_sq = d**2 + d.T ** 2
    # Implement the filter
    log = (1 - d_sq/2/(sigma**2)) * np.exp(-d_sq/2/(sigma**2)) / (sigma**2)
    return log

def get_local_maxima_3(data, threshold, sigmas, neighborhood_size=5):
    # See: https://stackoverflow.com/a/9113227/3672986
    data_region_max = filters.maximum_filter(data, neighborhood_size)
    maxima = (data == data_region_max)
    data_min = filters.minimum_filter(data, neighborhood_size)
    maxima[data < threshold] = 0

    labeled, num_objects = scipy.ndimage.label(maxima)
    slices = scipy.ndimage.find_objects(labeled)

    features = []
    x, y = [], []
    for dy, dx, dz in slices:
        x_center = int(round((dx.start + dx.stop - 1)/2))
        y_center = int(round((dy.start + dy.stop - 1)/2))
        z_center = int(round((dz.start + dz.stop - 1)/2))
        features.append({
            'x': x_center,
            'y': y_center,
            'radius': np.sqrt(2)*sigmas[z_center],
        })
        
    return features

def compute_multi_scale_features(image, sigmas, threshold, window_size=11):
    response = np.zeros(
        [image.shape[0], image.shape[1], sigmas.size]
    )
    for ii, sigma in enumerate(sigmas):
        w = get_LoG_filter(201, sigma=sigma)
        feature_response = scipy.signal.convolve(
            image, w, mode='same')
        response[:, :, ii] = np.abs(feature_response)
    
    return get_local_maxima_3(response, threshold, sigmas=sigmas,
                             neighborhood_size=window_size)

def plot_circ_features(image, features, ax):
    ax.imshow(image, cmap='gray')
    for m in features:
        #if m[2] == sigmas.max():
        #    continue
        cir = plt.Circle((m['x'], m['y']), m['radius'], color='r', fill=False)
        ax.add_artist(cir)
        
def compute_local_orientation(image, feature):
    sobel_x = np.array([
      [1, 0, -1],
      [2, 0, -2],
      [1, 0, -1]
    ])

    sobel_y = np.array([
      [1, 2, 1],
      [0, 0, 0],
      [-1, -2, -1]
    ])
    
    w = get_gaussian_filter(15, sigma=3)
    ir = feature['y']
    ic = feature['x']
    image_smoothed = scipy.signal.convolve(
        image, w, mode='same')
    Ix = scipy.signal.convolve(
        image_smoothed, sobel_x, mode='same')
    Iy = scipy.signal.convolve(
        image_smoothed, sobel_y, mode='same')
    return np.arctan2(Iy[ir, ic], Ix[ir, ic])
    
        
def compute_descriptors(image, features, patch_half_width=9):
    """Add descriptor patches (and orientation) to the features."""
    # Loop through the features
    for feature in features:
        # Blur the image
        scale = feature['radius']
        loc = [feature['y'], feature['x']]
        w = get_gaussian_filter(5*scale, scale/patch_half_width)
        scale_blurred_image = scipy.signal.convolve(
            image, w, mode='same')

        # Compute the orientation
        feature['orientation'] = compute_local_orientation(
            scale_blurred_image, feature)
    
        # Compute the rotated, scaled image patch
        patch = get_image_patch(
            scale_blurred_image,
            loc, 
            scale=scale,           
            orientation=feature['orientation'],
            patch_half_width=patch_half_width)
        patch = (patch - np.mean(patch))/np.std(patch)/np.prod(patch.shape)
        feature['patch'] = patch

    return features


fig = plt.figure()
im_half_size = 25
sigmas = np.arange(2, 20, 0.1)
circ_img_a = get_circ_image(2 * im_half_size + 1, radius=12)
circ_img_b = get_square_image(2 * im_half_size + 1, width=8)
circ_img = np.concatenate([circ_img_a, circ_img_b], axis=1)
plt.imshow(circ_img)
        
# Show for the circle image
image_base = circ_img
sigmas = np.arange(1.0, 10.0, 0.2)
features = compute_multi_scale_features(image_base, sigmas, 1.2)
features = compute_descriptors(image_base, features)
fig = plt.figure(figsize=(8, 8), dpi=300)
plot_circ_features(image_base, features, plt.gca())


```

```python
# Solution: Match Features (finally)
import cv2
res = cv2.resize(img, dsize=(54, 140), interpolation=cv2.INTER_CUBIC)

image_base = circ_img
image_base = load_image('light_cubes_base.png')[::2, ::2, 0]
image_base = load_image('sunflower_field.jpg')[:, :, 0]
sigmas = np.arange(5, 40.0, 1)
imagea = scipy.ndimage.rotate(image_base, 30)
fsa = compute_multi_scale_features(imagea, sigmas, 0.8)
fsa = compute_descriptors(imagea, fsa)

imageb = cv2.resize(image_base.T, 
                    dsize=tuple((np.array(image_base.shape)/1.5).astype(int)),
                    interpolation=cv2.INTER_CUBIC).T
imageb = scipy.ndimage.rotate(imageb, -45)
fsb = compute_multi_scale_features(imageb, sigmas, 0.8)
fsb = compute_descriptors(imageb, fsb)


def compute_feature_matches(fsa, fsb):
    # First compute the strength of the feature response
    sims = np.zeros((len(fsa), len(fsb)), dtype=np.float)
    for ii, fa in enumerate(fsa):
        for jj, fb in enumerate(fsb):
            sims[ii, jj] = np.sum(fa['patch'] * fb['patch'])
            
    print(sims)
    # Now compute the matches
    matches = []
    for ii in range(len(fsa)):
        mi = np.argmax(sims[ii])
        if not ii == np.argmax(sims[:, mi]):
            continue
        match_score = sims[ii, mi]
        matches.append([fsa[ii], fsb[mi]])
        
    return matches

def plot_matches(imagea, fsa, imageb, fsb, matches):
    s = np.maximum(imagea.shape, imageb.shape)
    img_a_pad = np.zeros(s)
    img_a_pad[:imagea.shape[0], :imagea.shape[1]] = imagea
    img_b_pad = np.zeros(s)
    img_b_pad[:imageb.shape[0], :imageb.shape[1]] = imageb
    img_concat = np.concatenate([img_a_pad, img_b_pad], axis=0)
    
    # First plot the features
    fig = plt.figure(figsize=(8, 8))
    plt.imshow(img_concat)

    plt.scatter([f['x'] for f in fsa], 
        [f['y'] for f in fsa], 
        s=4, facecolor="none", edgecolor='black')
    plt.scatter([f['x'] for f in fsb], 
        [f['y']+s[0] for f in fsb], 
        s=4, facecolor="none", edgecolor='black')
    
    # Now plot the matches:
    for m in matches:
        fa = m[0]
        fb = m[1]
        plt.plot([fa['x'], fb['x']],
                [fa['y'], fb['y'] + s[0]])


matches = compute_feature_matches(fsa, fsb)
plot_matches(imagea, fsa,
             imageb, fsb,
             matches)
```
